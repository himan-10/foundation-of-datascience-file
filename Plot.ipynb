```python
import matplotlib.pyplot as plt
x = [1, 3, 6, 8]
y = [2, 4, 5, 7]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=0,basefmt="g",label="python",use_line_collection=False)
plt.legend()
plt.title("stem plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define data
x = [10, 20, 30, 40, 50]
x1 = [20, 30, 40, 50, 60]
x2 = [30, 40, 50, 60, 70]
x3 = [35, 50, 60, 70, 80]
y = [x, x1, x2, x3]

# Create boxplots with labels
plt.boxplot(y, vert=False, patch_artist=True, labels=["python", "c++", "java", "javascript"])

# Add title, axis labels, and legend
plt.xlabel("x", fontsize=33)
plt.ylabel("y", fontsize=22)
plt.title("graph", fontsize=20)
plt.legend()
plt.show()
```
